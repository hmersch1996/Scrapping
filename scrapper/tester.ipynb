```python
from utils import *
from datetime import datetime
#from sqlalchemy import create_engine
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
URL = 'https://www.tigo.com.py/catalogo-prepago'

TODAY = datetime.today()
driver = initChrome()
driver.get(URL)
df=pd.DataFrame(columns=["nombre","precio_total","cuotas","precio_cuotas","financiamiento","img","url","fecha_extraccion"])

items=driver.find_elements(By.CLASS_NAME,"column.mb-2.is-6-tablet.is-4-fullhd")
for item in items:
    marca=item.find_element(By.TAG_NAME,"h3").text
    modelo=item.find_element(By.TAG_NAME,"h2").text
    title = marca + " " + modelo

    price = item.find_element(By.CLASS_NAME,"is-size-6.text-secondary").text
    new_price=(price.replace('.','')).replace("Gs","")

    # extraemos la imagen y el link de referencia, junto con el costo de las cuotas
    img = item.find_element(By.TAG_NAME,"img").get_attribute("src")
    link = item.find_element(By.TAG_NAME,"a").get_attribute("href")
    fee_price = new_price
    fees = 1
    df.loc[len(df),:]=[title,new_price,fees,fee_price,"Contado",img,link,TODAY]

print("Se ha completado la extraccion de datos")

sleep(10)
driver.close()
```
